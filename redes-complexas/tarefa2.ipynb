```python
import igraph as ig
import numpy as np
import matplotlib.pyplot as plt
```

## Atividades Propostas

1. Carregue a rede do Karatê: 

```python
g = ig.Graph.Read_GML('karate.gml')
```

2. Calcule e imprima: N, L, <k>, L_{max}, d_{max} (número de nós, número de links, grau médio, número máximo possível de arestas e diâmetro):

```python
n_nos = g.vcount()
n_links = g.ecount()
grau_medio = np.mean(g.degree())
l_max = (n_nos * (n_nos - 1.0))/2.0
diametro = g.diameter()

print(f'N - Número de nós: {n_nos}')
print(f'L - Número de links: {n_links}')
print(f'<k> - Grau médio: {(grau_medio): .2f}')
print(f'L_max - Número máximo possível de arestas: {l_max}')
print(f'd_max - Diâmetro: {diametro}')
```

3. Gere um plot da rede: 


```python
ig.plot(g)
```

4. Gere um plot de p_k (distribuição dos graus):

```python
graus = g.degree()

plt.hist(graus, 
         bins=range(min(graus), max(graus) + 2), 
         align='left', 
         rwidth=0.8, 
         alpha=0.7,
         density= True
         )


plt.xticks(range(min(graus), max(graus) +1))
plt.ylim(0,1)
plt.xlabel('Grau')
plt.ylabel('Densidade dos nós')
plt.title('Distribuição dos Graus do Grafo')

```
