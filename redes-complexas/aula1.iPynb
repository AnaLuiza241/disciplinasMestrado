```python
import igraph as ig
import networkx as nx
import matplotlib.pyplot as plt
```

```python
#  Criando e plotando um grafo usando Networkx

G = nx.Graph()
G.add_nodes_from([1,2,3,4])
G.add_edges_from([(1,2),(1,3),(2,4),(2,3)])
print(f'Números de Nós:{G.number_of_nodes()}') 
print(f'Números de Arestas:{G.number_of_edges()}')

labels = {1: '1', 2: '2', 3: '3', 4: '4'}

nx.draw(G, with_labels=True, labels= labels)
```

```python
# Criando e plotando um grafo usando o igraph

G = ig.Graph()
G.add_vertices(4)
G.add_edges([(0,2), (0,1), (1,2), (1,3)])
print(f'Números de Nós:{G.vcount()}') 
print(f'Números de Arestas:{G.ecount()}')

G.vs['label'] = ['1', '2', '3', '4 ']

ig.plot(G, vertex_label=G.vs['label'])
```
