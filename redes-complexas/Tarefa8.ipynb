```python
import networkx as nx
import matplotlib.pyplot as plt
import community as community_louvain
```

```python
g = nx.read_edgelist('fb-pages-food.edges', delimiter=',')
#g = nx.read_edgelist('/content/drive/MyDrive/fb-pages-food.edges', delimiter=',')

# Gerar layout para a rede (usando o layout 'spring' - similar ao Fruchterman-Reingold)
pos = nx.spring_layout(g)

# Plotar a rede
plt.figure(figsize=(12, 10))
nx.draw(
    g, pos,
    node_size=20,            # Tamanho dos nós
    node_color='skyblue',    # Cor dos nós
    edge_color='gray',       # Cor das arestas
    width=0.5,               # Espessura das arestas
    with_labels=False        # Não mostrar rótulos dos nós
)

# Exibir o gráfico
plt.title("Visualização da Rede - fb-pages-food")
plt.show()
```

```python
import networkx as nx
import community.community_louvain  # Importação correta para o Louvain
import matplotlib.pyplot as plt

# Carregar a rede
g = nx.read_edgelist('/content/drive/MyDrive/fb-pages-food.edges', delimiter=',')

# Detectando as comunidades usando o algoritmo Louvain
partition = community.community_louvain.best_partition(g)

# Calculando a modularidade
modularity = community.community_louvain.modularity(partition, g)
print(f'Modularidade: {modularity}')

# Plotando o grafo com cores diferentes para as comunidades
plt.figure(figsize=(12, 10))

# Atribuindo cores diferentes para cada comunidade
colors = [partition[node] for node in g.nodes()]

# Desenhando o grafo
nx.draw(
    g, node_color=colors, with_labels=False, node_size=20,
    cmap=plt.cm.jet, edge_color='gray', width=0.5
)

# Título do gráfico
plt.title("Comunidades no Grafo com Modularidade")
plt.show()
```
